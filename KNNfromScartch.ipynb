```python
import numpy as np
import pandas as pd
import math
```

```python


points = np.array([
    [1,4],
    [2,8],
    [4,1],
    [9,4],
    [6,7],
    [9,1],
    [8,9],
    [6,3]
])

L = ["A","B","A","A","B","B","A","B"]

labels = np.array(L)

print(points)
print(labels)
```

```python
k=int(input("Enter value for K:"))
Pt=input("Enter points:")
Pt=Pt.split(",")
X=int(Pt[0])
Y=int(Pt[1])
a=[X,Y]
pt=np.array(a)
print(pt)
```

```python
import math

def dist(a, b):
    s = 0
    for i in range(len(a)):
        s += (a[i] - b[i]) ** 2
    return math.sqrt(s)

```

```python
d=[]
for i in range(7):
   d.append(dist(points[i],pt))
print(d)
```

```python
pairs = []

for i in range(7):
    pairs.append((d[i], labels[i]))

pairs.sort(key=lambda x: x[0])

print(pairs)
```

```python
from collections import Counter
nearest = pairs[:k]


# Step 4: Majority vote
nearest_labels = [label for _, label in nearest]

prediction = Counter(nearest_labels).most_common(1)[0][0]


print("Nearest Neighbors:", nearest)
print("Predicted Class:", prediction)
```
